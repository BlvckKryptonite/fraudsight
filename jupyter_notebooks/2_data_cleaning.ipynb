```python
import pandas as pd
import os

# So I can check working directory for debugging
print("Current Working Directory:", os.getcwd())

# Corrected dataset path
df = pd.read_csv("../data/transactions.csv")

# Display basic info to verify loading
print("Dataset Shape:", df.shape)
print("\nFirst 5 Rows:")
print(df.head())
print("\nMissing Values:")
print(df.isnull().sum())

# Handle missing values
df.dropna(inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Saving the cleaned data 
os.makedirs("../data", exist_ok=True)
df.to_csv("../data/cleaned_transactions.csv", index=False)
print("\nCleaned dataset saved to '../data/cleaned_transactions.csv'")
```
