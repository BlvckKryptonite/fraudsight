```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# dataset path
df = pd.read_csv("../data/cleaned_transactions.csv")

# Standardize fraud label
df = df.rename(columns={"Class": "Is_Fraud"})

# Create Hour column from Time 
df["Hour"] = pd.to_datetime(df["Time"], unit="s").dt.hour

df.head()

# Fraud rate by hour
plt.figure(figsize=(12, 6))
sns.barplot(x="Hour", y="Is_Fraud", data=df)
plt.title("Fraud Rate by Hour")
plt.show()

# Transaction amount distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x="Is_Fraud", y="Amount", data=df)
plt.title("Transaction Amount by Fraud Status")
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# User transaction frequency
plt.figure(figsize=(12, 6))
sns.countplot(x="User_ID", hue="Is_Fraud", data=df)
plt.title("Transactions per User (Fraud vs. Legitimate)")
plt.show()
```
